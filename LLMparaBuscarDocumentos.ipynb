```python
#listagem de documentos que serão buscados
DOCUMENT1 = {
    "title": "Introdução à Inteligência Artificial",
    "conteudo": "A inteligência artificial (IA) é uma área da ciência da computação dedicada à criação de sistemas que realizam tarefas que normalmente requerem inteligência humana. Exemplos incluem reconhecimento de voz, aprendizado de máquina e sistemas de recomendação."
}

DOCUMENT2 = {
    "title": "Explorando a Internet das Coisas (IoT)",
    "conteudo": "A Internet das Coisas (IoT) refere-se à interconexão de dispositivos através da internet, permitindo que eles coletem e compartilhem dados. Isso pode incluir desde eletrodomésticos até sistemas industriais, criando uma rede integrada de informações."
}

DOCUMENT3 = {
    "title": "O Futuro da Computação Quântica",
    "conteudo": "A computação quântica utiliza princípios da mecânica quântica para processar informações de maneira muito mais eficiente do que os computadores clássicos. Com potencial para revolucionar áreas como criptografia e simulação de moléculas, a computação quântica está na vanguarda da inovação tecnológica."
}

documents = [DOCUMENT1, DOCUMENT2, DOCUMENT3]

```

```python
df = pd.DataFrame(documents)
df.columns = ['title', 'conteudo']
df
```

```python
model = "models/embedding-001"
```

```python
def get_embedding(title, text):
  return genai.embed_content(model=model,
                             content=text,
                             title=title,
                             task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["embedding"] = df.apply(lambda row: get_embedding(row.title, row.conteudo), axis=1)
df
```
